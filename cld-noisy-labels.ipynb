```python
WORK_DIR = '../input/cassava-leaf-disease-classification'
TARGET_SIZE = 512
```

```python
model = keras.models.load_model('../input/cldbestsinglemodels/CLD - Best models/0.883 - effnetB3 single_model/EfNetB3_untrained_TPU_model (1).h5')
model.load_weights('../input/cldbestsinglemodels/CLD - Best models/0.883 - effnetB3 single_model/EffNetB3_TPU_Model_best_weights (1).h5')
```

```python
train_df = pd.read_csv('../input/cassava-leaf-disease-classification/train.csv')
train_df.head(3)
```

```python
#making preds on all the test_images
preds = []
confidence = []

for image_id in train_df.image_id:
    image = Image.open(os.path.join(WORK_DIR,"train_images", image_id))
    image = image.resize((TARGET_SIZE, TARGET_SIZE))
    image = np.expand_dims(image, axis = 0)
    image = image/255 #normalizing pixel vals before predicting
    preds.append(np.argmax(model.predict(image)))
    confidence.append(np.max(model.predict(image)))

train_df['pred_label'] = preds
train_df['confidence'] = confidence

train_df.to_csv('submission.csv', index = False)
```
